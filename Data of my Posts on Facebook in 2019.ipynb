```python
dt = pd.read_json('D:/Data/Files/JSON/facebook-kongsansrip/posts/your_posts_1.json')
dt
```

```python
dt['date'] = ts.dt.date
dt
```

```python
ax = dt['date'].value_counts().plot(figsize=(20,5))
ax.set(title='Timeline of my posts in 2019', xlabel='Month', ylabel='Number');
```

```python
ax2 = dt['date'].value_counts().head(10).plot.bar(figsize=(10,5), rot=45)
ax2.set(title='Top Ten Highest posts per day on my Facebook in 2019', xlabel='Date', ylabel='Number');
```
