```python
df = pd.read_json('D:/Data/Files/JSON/facebook-kongsansrip/likes_and_reactions/posts_and_comments_edited.json')
df
```

```python
ts = df['timestamp']
ts
```

```python
df['date'] = ts.dt.date
df['date']
```

```python
ax1 = df['date'].value_counts().plot(figsize=(20,5));
ax1.set(title='Timeline of my Reactions on Facebook in 2019', xlabel='Month', ylabel='Number');
```

```python
df['date'].value_counts()
```

```python
df['data'].value_counts()
```

```python
reaction = []
for datas in df['data']:
    reaction.append(datas[0]['reaction'])
df_g = pd.DataFrame([reaction]).T
```

```python
reaction = []
for react in df_g[0]:
    reaction.append(react['reaction'])
df_s = pd.DataFrame([reaction]).T
```

```python
df_s
```

```python
df['reaction'] = df_s
```

```python
df
```

```python
ax2 = df['reaction'].value_counts().sort_values().plot.bar(rot=0);
ax2.set(title='Number of my Reactions on Facebook in 2019 ', xlabel='Reaction Type', ylabel='Number');
```

```python
df['time'] = ts.dt.time
df
```

```python
dtt = df['title'].str.replace('Kor\sSriprasit','')
dtt
```

```python
dtt1 = dtt.str.replace('(likes\s)|(reacted\sto\s)','')
dtt1
```

```python
dtt2 = dtt1.str.replace('(\'s\sphoto\.)|(\'s\spost\.)|(\'s\svideo\.)','')
dtt2
```

```python
df['user'] = dtt2
```

```python
df
```

```python
dfr = df[['date','time','reaction','user']]
dfr
```
